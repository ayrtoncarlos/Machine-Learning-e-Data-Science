```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salarios_ayrton = np.array([100, 200, 300, 400])
salarios_cristine = np.array([1000, 2000, 3000, 4000])
```

```python
# Adicionando Label
plt.plot(salarios_ayrton, c="Red", ls="--", marker="s", ms=10, label="Sal치rios Ayrton")
plt.plot(salarios_cristine, c="green", ls="--", marker="s", ms=10, label="Sal치rios Cristine")
plt.show()
```

```python
# Colocando Legenda
plt.plot(salarios_ayrton, c="Red", ls="--", marker="s", ms=10, label="Sal치rios Ayrton")
plt.plot(salarios_cristine, c="green", ls="--", marker="s", ms=10, label="Sal치rios Cristine")
plt.legend()
plt.show()
```

```python
# Modificando Marcador
plt.plot(salarios_ayrton, c="Red", ls="--", marker="s", ms=10, label="Sal치rios Ayrton")
plt.plot(salarios_cristine, c="green", ls="--", marker="^", ms=10, label="Sal치rios Cristine")
plt.legend()
plt.show()
```

```python
# Modificar local onde legenda esta inserida
plt.plot(salarios_ayrton, c="Red", ls="--", marker="s", ms=10, label="Sal치rios Ayrton")
plt.plot(salarios_cristine, c="green", ls="--", marker="s", ms=10, label="Sal치rios Cristine")
plt.legend(loc="center right")
plt.show()
```

```python
# Colocando mais uma linha
salarios_rafaela = [500, 400, 200, 300]
plt.plot(salarios_ayrton, c="Red", ls="--", marker="s", ms=10, label="Sal치rios Ayrton")
plt.plot(salarios_cristine, c="Green", ls="--", marker="s", ms=10, label="Sal치rios Cristine")
plt.plot(salarios_rafaela, c="Blue", ls="--", marker="o", ms=10, label="Sal치rios Rafaela")
plt.legend(loc="upper left")
plt.show()
```
