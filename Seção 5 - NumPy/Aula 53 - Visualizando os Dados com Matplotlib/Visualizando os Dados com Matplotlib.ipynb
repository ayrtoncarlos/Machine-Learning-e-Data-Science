```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
salario = np.array([200, 300, 400, 150, 100, 500])
salario
```

```python
plt.plot(salario)
```

```python
plt.plot(salario, c="Red")
```

```python
plt.plot(salario, c="Red", ls="--", marker="s")
```

```python
plt.plot(salario, c="Red", ls="--", marker="s", ms=10)
plt.show()
```
