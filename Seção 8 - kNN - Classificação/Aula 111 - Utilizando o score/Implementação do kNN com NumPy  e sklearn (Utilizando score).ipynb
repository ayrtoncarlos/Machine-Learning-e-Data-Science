```python
"""
Dataset: https://archive.ics.uci.edu/ml/datasets/Balance+Scale

Attribute Information:

1. Class Name: 3 (L, B, R) 
2. Left-Weight: 5 (1, 2, 3, 4, 5) 
3. Left-Distance: 5 (1, 2, 3, 4, 5) 
4. Right-Weight: 5 (1, 2, 3, 4, 5) 
5. Right-Distance: 5 (1, 2, 3, 4, 5)

São 625 exemplos, 4 atributos e 3 possíveis classes.
Onde tinha 'L' coloquei 1, onde tinha 'B' coloquei 2, e onde tinha 'R' coloquei 3.
"""
```

```python
import numpy as np
```

```python
# x são as entradas e y são as saídas

x = np.genfromtxt("dataset.data", delimiter=",", usecols=(1, 2, 3, 4))
y = np.genfromtxt("dataset.data", delimiter=",", usecols=(0))

print(len(x))
print(len(y))
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Doc: http://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(len(x_treino))
x_treino
```

```python
print(len(x_teste))
x_teste
```

```python
print(len(y_treino))
y_treino
```

```python
print(len(y_teste))
y_teste
```

```python
y_teste[0]
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=17, p=2)
knn.fit(x_treino, y_treino)
labels = knn.predict(x_teste)
print(len(labels))
print(labels)
```

```python
y_teste
```

```python
np.sum(labels == y_teste)
```

```python
(labels == y_teste).sum()
```

```python
100 * (labels == y_teste).sum() / len(x_teste)
```

```python
knn.score(x_teste, y_teste)
```
