```python
%matplotlib inline
%config InlineBackend.figure_format='svg'
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("reported.csv")
```

```python
df.head()
```

```python
df.fillna(0, inplace=True)
```

```python
df.head()
```

```python
plt.plot(df["Year"], df["crimes.total"], "-r")
plt.plot(df["Year"], df["crimes.person"], "-g")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["Year"], df["crimes.total"], "-r")
ax.plot(df["Year"], df["crimes.person"], "-g")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["Year"], df["crimes.total"], "-r")
ax.plot(df["Year"], df["crimes.person"], "-g")
ax.legend()
ax.set_xlabel("Year")
ax.set_ylabel("Quantity")
ax.set_title("Crimes: Total x Person")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(df["Year"], df["crimes.total"], "-r")
ax.plot(df["Year"], df["crimes.person"], "-g")
ax.legend()
ax.set_xlabel("Year")
ax.set_ylabel("Quantity")
ax.set_title("Crimes: Total x Person")
ax.set_xlim([df["Year"].min(), df["Year"].max()])
plt.show()
```
