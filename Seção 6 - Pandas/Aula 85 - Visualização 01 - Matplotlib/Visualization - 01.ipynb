```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x = np.linspace(1, 10, 100)
```

```python
x
```

```python
plt.plot(x, np.sin(x))
plt.show()
```

```python
plt.plot(x, np.sin(x), "r")
plt.show()
```

```python
plt.plot(x, np.sin(x), "--")
plt.show()
```

```python
plt.plot(x, np.sin(x), "r--")
plt.show()
```

```python
fig = plt.figure()
plt.plot(x, np.sin(x), "r--")
plt.show()
```
