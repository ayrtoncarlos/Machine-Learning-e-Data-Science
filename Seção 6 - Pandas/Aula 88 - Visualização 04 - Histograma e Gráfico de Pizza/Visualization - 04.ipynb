```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = pd.read_csv("Titanic.csv", delimiter=",")
```

```python
titanic.head()
```

```python
titanic["PClass"].value_counts()
```

```python
titanic["PClass"].value_counts().plot(kind="bar")
plt.show()
```

```python
titanic["Survived"].value_counts().plot(kind="bar")
plt.show()
```

```python
titanic.groupby("Survived")["PClass"].value_counts()
```

```python
titanic.groupby("Survived")["PClass"].value_counts().plot(kind="bar")
plt.show()
```

```python
estados = pd.read_csv("populacao_brasil.csv")
```

```python
estados.head()
```

```python
estados["total"].hist()
plt.show()
```

```python
fig, ax = plt.subplots()
plt.hist(estados["total"], bins=15, orientation="horizontal")
ax.ticklabel_format(style="plain")
plt.show()
```

```python
estados["percent"] = estados["total"] / estados["total"].sum()
```

```python
estados.head()
```

```python
%config InlineBackend.figure_format='svg'
```

```python
plt.pie(estados["percent"], labels=estados["estado"])
plt.show()
```

```python
plt.pie(estados["percent"], labels=estados["estado"], autopct="%1.2f%%")
plt.show()
```
